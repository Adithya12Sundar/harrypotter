```python
import random
import matplotlib.pyplot as plt

total_simulations = int(input("How many times should I simulat"))
faika = []
for i in range(total_simulations):
  faika.append(random.randint(1,6))
print(faika)
plt.hist(bins = 6, x = faika)
plt.show()
```
